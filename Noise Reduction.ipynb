```python
import imageio
from scipy.signal import convolve2d
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python

img = cv2.imread('gaussian.png')

dst = cv2.fastNlMeansDenoisingColored(img,None,10,10,7,21)

plt.subplot(121),plt.imshow(img)
plt.subplot(122),plt.imshow(dst)
plt.show()
```

```python
# average filter w/ salt n pepper noise

img = imageio.imread('noisy-img1.png')
imgd = img.astype(float) / 255.0  # normalize to [0, 1]
imgd = np.clip(imgd + 0.02 * (np.random.rand(*imgd.shape) - 0.5), 0, 1)  # add salt-and-pepper noise
f = np.ones((3, 3)) / 9
img1 = convolve2d(imgd, f, mode='same')
plt.subplot(121)
plt.imshow(imgd, cmap='gray')
plt.subplot(122)
plt.imshow(img1, cmap='gray')
plt.show()
```
